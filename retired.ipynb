```python
###General MWC model
def sugar_concentration(positions):
    "Calculate the sugar concentration based on position."
    return np.sqrt(positions[0]**2 + positions[1]**2)

def fraction_R(L, S, K_active, K_inactive, n):
    "Calculate the fraction of receptors in the R state based on the MWC model."
    numerator = L * (S / K_active)**n
    denominator = numerator + (L* (1 + (K_active/ K_inactive)*(S/K_active))**n)
    return numerator / denominator

def compute_tumble_probability(L, n, K_active, K_inactive, position, previous_position):
    "Calculate the probability of tumbling"
    current_concentration = sugar_concentration(position)
    previous_concentration = sugar_concentration(previous_position)
    active_fraction_current = fraction_R(L, current_concentration, K_active, K_inactive, n)
    active_fraction_previous = fraction_R(L, previous_concentration, K_active, K_inactive, n)
    active_fraction_difference = np.clip(active_fraction_current - active_fraction_previous, -0.7, 0.3) ###possibly omit negative
    # A negative change increases tumble probability, positive change reduces it
    return 0.3 - active_fraction_difference

def choose_new_direction():
    "Choose a new random direction."
    direction = np.random.uniform(0, 2*np.pi)
    return direction

def simulate_movement(duration, L, n, K_active, K_inactive, initial_position, step_size=0.5):
    "Simulate bacterial movement"
    positions = [np.array(initial_position)]
    sugar_concentrations = [sugar_concentration(initial_position)]  
    time_step = 0.1
    steps = int(duration / time_step)  # Calculate the number of steps based on the time step
    direction = np.pi/4

    for _ in range(steps):
        position = positions[-1]
        if len(positions) > 1:
            tumble_probability = compute_tumble_probability(L, n, K_active, K_inactive, position, positions[-2])
        else:
            tumble_probability = 0.99  # High initial tumble probability to simulate initial exploration

        if np.random.rand() < tumble_probability:
            direction = choose_new_direction()

        new_position = position + step_size * np.array([np.cos(direction), np.sin(direction)])
        positions.append(new_position)
        sugar_concentrations.append(sugar_concentration(new_position))  # Record the sugar concentration at new position

    return np.array(positions), sugar_concentrations

# Parameters
### S = the ligand (sugar) concentration
L = 10  # Equilibrium constant between R and T states without ligand
n = 5  # Number of binding sites
K_active = 1  # Dissociation constant for active state
K_inactive = 10  # Dissociation constant for inactive state
duration = 500  # Simulation duration in arbitrary units
initial_position = [0, 0]  # Starting position

# Run simulation
positions, sugar_concentrations = simulate_movement(duration, L, n, K_active, K_inactive, initial_position)

plt.figure(figsize=(10, 6))
# Generate a mesh grid for the background concentration levels
x = np.linspace(min(positions[:, 0]) - 1, max(positions[:, 0]) + 1, 400)
y = np.linspace(min(positions[:, 1]) - 1, max(positions[:, 1]) + 1, 400)
X, Y = np.meshgrid(x, y)
Z = sugar_concentration([X, Y]) 

plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')
plt.colorbar(label='Sugar Concentration')
plt.plot(positions[:, 0], positions[:, 1], marker='o', markersize=0.1, color='red')
plt.title('Bacterial Movement Over Time on a Spatial Sugar Gradient with MWC Model for receptor activation')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.axis('equal')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(sugar_concentrations, marker='o', linestyle='-', markersize=0.1, color='blue')
plt.title('Sugar Concentration Experienced by Bacterium Over Time')
plt.xlabel('Time Step')
plt.ylabel('Sugar Concentration')
plt.grid(True)
plt.show()
```

```python
def joint_prob(x, y, num_bins):
    joint_hist, x_edges, y_edges = np.histogram2d(x, y, bins=num_bins)
    joint_prob = joint_hist / np.sum(joint_hist)
    return joint_prob, x_edges, y_edges 

def mutual_information(x, y, num_bins):
    p_xy, x_edges, y_edges = joint_prob(x, y, num_bins)
    p_x = np.sum(p_xy, axis=1, keepdims=True)
    p_y = np.sum(p_xy, axis=0, keepdims=True)

    # Avoid log of zero and division by zero
    non_zero = p_xy > 0
    p_xy_nonzero = p_xy[non_zero]
    
    p_x_nonzero = p_x[p_x > 0].flatten()
    p_y_nonzero = p_y[p_y > 0].flatten()
    p_x_y_nonzero = np.outer(p_x_nonzero, p_y_nonzero)
    
    p_x_y_nonzero = p_x_y_nonzero.flatten()[non_zero.flatten()]

    mi = np.sum(p_xy_nonzero * np.log(p_xy_nonzero / p_x_y_nonzero))
    return mi

# Function to compute MI over time with a sliding window
def compute_mi_over_time(c, a, m, window_size, num_bins):
    mi_c_a = []
    mi_c_m = []
    mi_a_m = []

    for i in range(len(c) - window_size + 1):
        c_window = c[i:i + window_size]
        a_window = a[i:i + window_size]
        m_window = m[i:i + window_size]

        mi_c_a.append(mutual_information(c_window, a_window, num_bins))
        mi_c_m.append(mutual_information(c_window, m_window, num_bins))
        mi_a_m.append(mutual_information(a_window, m_window, num_bins))

    return mi_c_a, mi_c_m, mi_a_m

def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

window_size = 100  # window size for MI calculation
num_bins = 5      # Number of bins for discretization
smoothing_window = 100  #window size for MA

# Compute MI
mi_c_a, mi_c_m, mi_a_m = compute_mi_over_time(sugar_concentrations, CheA_activities, methylation_levels, window_size, num_bins)

# Apply moving average smoothing
mi_c_a_smooth = moving_average(mi_c_a, smoothing_window)
mi_c_m_smooth = moving_average(mi_c_m, smoothing_window)
mi_a_m_smooth = moving_average(mi_a_m, smoothing_window)
smooth_time = time[:len(mi_c_a_smooth)]

mi_df = pd.DataFrame({
    'Time': smooth_time,
    'MI(c; a)': mi_c_a_smooth,
    'MI(c; m)': mi_c_m_smooth,
    'MI(a; m)': mi_a_m_smooth
})

plt.figure(figsize=(14, 6))
plt.plot(mi_df['Time'], mi_df['MI(c; a)'], label='MI(c; a)', color='blue')
plt.plot(mi_df['Time'], mi_df['MI(c; m)'], label='MI(c; m)', color='green')
plt.plot(mi_df['Time'], mi_df['MI(a; m)'], label='MI(a; m)', color='red')
plt.xlabel('Time Step')
plt.ylabel('Mutual Information')
plt.title('Smoothed Dynamic Mutual Information Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def compute_entropy(prob_dist):
    nzs = prob_dist > 0  # Non-zero probabilities
    entropy = -np.sum(prob_dist[nzs] * np.log(prob_dist[nzs]))
    return entropy

def compute_histogram(time_series_1, time_series_2, bins):
    hist_2d, x_edges, y_edges = np.histogram2d(time_series_1, time_series_2, bins=bins)
    pxy = hist_2d / float(np.sum(hist_2d))
    px = np.sum(pxy, axis=1)
    py = np.sum(pxy, axis=0)
    return pxy, px, py

def compute_mutual_information_and_entropies(time_series_1, time_series_2, bins=30):
    pxy, px, py = compute_histogram(time_series_1, time_series_2, bins)
    nzs = pxy > 0  # Non-zero probabilities
    
    mi = np.sum(pxy[nzs] * np.log(pxy[nzs] / (px[:, None] * py[None, :])[nzs]))
    
    H_x = compute_entropy(px)
    H_y = compute_entropy(py)
    H_xy = compute_entropy(pxy.flatten())
    H_x_given_y = H_xy - H_y
    
    return mi, H_x, H_x_given_y

def compute_marginal_entropies(all_sugar_concentrations, all_methylation_levels, all_CheY_levels, all_tumbling_probabilities, num_runs, dt, bins=30):
    num_steps = len(all_sugar_concentrations[0])
    
    H_sugar = []
    H_methylation = []
    H_CheY = []
    H_tumbling = []

    for step in range(num_steps):
        sugar_at_step = [all_sugar_concentrations[run][step] for run in range(num_runs)]
        methylation_at_step = [all_methylation_levels[run][step] for run in range(num_runs)]
        CheY_at_step = [all_CheY_levels[run][step] for run in range(num_runs)]
        tumbling_at_step = [all_tumbling_probabilities[run][step] for run in range(num_runs)]
        
        _, px, _ = compute_histogram(sugar_at_step, sugar_at_step, bins)
        H_sugar.append(compute_entropy(px))
        
        _, px, _ = compute_histogram(methylation_at_step, methylation_at_step, bins)
        H_methylation.append(compute_entropy(px))
        
        _, px, _ = compute_histogram(CheY_at_step, CheY_at_step, bins)
        H_CheY.append(compute_entropy(px))
        
        _, px, _ = compute_histogram(tumbling_at_step, tumbling_at_step, bins)
        H_tumbling.append(compute_entropy(px))

    return H_sugar, H_methylation, H_CheY, H_tumbling

def compute_instantaneous_mutual_information_with_derivative_and_entropies(all_sugar_derivatives, all_methylation_levels, all_CheY_levels, all_tumbling_probabilities, num_runs, dt, bins=30):
    num_steps = len(all_sugar_derivatives[0])
    
    mi_sugar_derivative_methylation = []
    H_sugar_derivative = []
    H_sugar_derivative_given_methylation = []
    
    mi_sugar_derivative_CheY = []
    H_sugar_derivative_given_CheY = []
    
    mi_sugar_derivative_tumbling = []
    H_sugar_derivative_given_tumbling = []

    for step in range(num_steps):
        sugar_derivative_at_step = [all_sugar_derivatives[run][step] for run in range(num_runs)]
        
        # Methylation
        methylation_at_step = [all_methylation_levels[run][step] for run in range(num_runs)]
        mi, H_x, H_x_given_y = compute_mutual_information_and_entropies(sugar_derivative_at_step, methylation_at_step, bins)
        mi_sugar_derivative_methylation.append(mi)
        H_sugar_derivative.append(H_x)
        H_sugar_derivative_given_methylation.append(H_x_given_y)
        
        # CheY-P
        CheY_at_step = [all_CheY_levels[run][step] for run in range(num_runs)]
        mi, _, H_x_given_y = compute_mutual_information_and_entropies(sugar_derivative_at_step, CheY_at_step, bins)
        mi_sugar_derivative_CheY.append(mi)
        H_sugar_derivative_given_CheY.append(H_x_given_y)
        
        # Tumbling probability
        tumbling_at_step = [all_tumbling_probabilities[run][step] for run in range(num_runs)]
        mi, _, H_x_given_y = compute_mutual_information_and_entropies(sugar_derivative_at_step, tumbling_at_step, bins)
        mi_sugar_derivative_tumbling.append(mi)
        H_sugar_derivative_given_tumbling.append(H_x_given_y)

    return (mi_sugar_derivative_methylation, H_sugar_derivative, H_sugar_derivative_given_methylation,
            mi_sugar_derivative_CheY, H_sugar_derivative_given_CheY,
            mi_sugar_derivative_tumbling, H_sugar_derivative_given_tumbling)

def moving_average(data, window_size):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')

num_runs = 1000
duration = 500
dt = 0.01
time = np.arange(0, duration, dt)
nbins = 30

# Compute derivatives of sugar concentrations
all_sugar_derivatives = []
for sugar_concentration in all_sugar_concentrations:
    sugar_derivative = np.gradient(sugar_concentration, dt)
    all_sugar_derivatives.append(sugar_derivative)

results = compute_instantaneous_mutual_information_with_derivative_and_entropies(
    all_sugar_derivatives, all_methylation_levels, all_CheY_levels, all_tumbling_probabilities, num_runs, dt
)

(mi_sugar_derivative_methylation, H_sugar_derivative, H_sugar_derivative_given_methylation,
 mi_sugar_derivative_CheY, H_sugar_derivative_given_CheY,
 mi_sugar_derivative_tumbling, H_sugar_derivative_given_tumbling) = results

H_sugar, H_methylation, H_CheY, H_tumbling = compute_marginal_entropies(
    all_sugar_concentrations, all_methylation_levels, all_CheY_levels, all_tumbling_probabilities, num_runs, dt
)

# Apply smoothing
window_size = 100
mi_sugar_methylation_smooth = moving_average(mi_sugar_derivative_methylation, window_size)
H_sugar_derivative_smooth = moving_average(H_sugar_derivative, window_size)
H_sugar_derivative_given_methylation_smooth = moving_average(H_sugar_derivative_given_methylation, window_size)

mi_sugar_derivative_CheY_smooth = moving_average(mi_sugar_derivative_CheY, window_size)
H_sugar_derivative_given_CheY_smooth = moving_average(H_sugar_derivative_given_CheY, window_size)

mi_sugar_derivative_tumbling_smooth = moving_average(mi_sugar_derivative_tumbling, window_size)
H_sugar_derivative_given_tumbling_smooth = moving_average(H_sugar_derivative_given_tumbling, window_size)

H_sugar_smooth = moving_average(H_sugar, window_size)
H_methylation_smooth = moving_average(H_methylation, window_size)
H_CheY_smooth = moving_average(H_CheY, window_size)
H_tumbling_smooth = moving_average(H_tumbling, window_size)

H_sugar_given_methylation_smooth = moving_average(H_sugar_derivative_given_methylation, window_size)
H_sugar_given_CheY_smooth = moving_average(H_sugar_derivative_given_CheY, window_size)
H_sugar_given_tumbling_smooth = moving_average(H_sugar_derivative_given_tumbling, window_size)

time_smooth = time[:len(mi_sugar_methylation_smooth)]

fig, axs = plt.subplots(2, 2, figsize=(16, 12))

axs[0, 0].plot(time_smooth, mi_sugar_methylation_smooth, label='MI(Sugar, Methylation)')
axs[0, 0].plot(time_smooth, mi_sugar_CheY_smooth, label='MI(Sugar, CheY-P)')
axs[0, 0].plot(time_smooth, mi_sugar_tumbling_smooth, label='MI(Sugar, Tumbling Probability)')
axs[0, 0].set_xlabel('Time (s)')
axs[0, 0].set_ylabel('Mutual Information (nats)')
axs[0, 0].set_title('Mutual Information between Sugar and Variables')
axs[0, 0].legend()

axs[0, 1].plot(time_smooth, mi_sugar_derivative_methylation_smooth, label='MI(Sugar Gradient, Methylation)')
axs[0, 1].plot(time_smooth, mi_sugar_derivative_CheY_smooth, label='MI(Sugar Gradient, CheY-P)')
axs[0, 1].plot(time_smooth, mi_sugar_derivative_tumbling_smooth, label='MI(Sugar Gradient, Tumbling Probability)')
axs[0, 1].set_xlabel('Time (s)')
axs[0, 1].set_ylabel('Mutual Information (nats)')
axs[0, 1].set_title('Mutual Information between Sugar Gradient and Variables')
axs[0, 1].legend()

axs[1, 0].plot(time[:len(H_sugar_smooth)], H_sugar_smooth, label='H(Sugar)')
axs[1, 0].plot(time[:len(H_methylation_smooth)], H_methylation_smooth, label='H(Methylation)')
axs[1, 0].plot(time[:len(H_CheY_smooth)], H_CheY_smooth, label='H(CheY-P)')
axs[1, 0].plot(time[:len(H_tumbling_smooth)], H_tumbling_smooth, label='H(Tumbling Probability)')
axs[1, 0].plot(time[:len(H_sugar_given_methylation_smooth)], H_sugar_given_methylation_smooth, label='H(Sugar | Methylation)')
axs[1, 0].plot(time[:len(H_sugar_given_CheY_smooth)], H_sugar_given_CheY_smooth, label='H(Sugar | CheY-P)')
axs[1, 0].plot(time[:len(H_sugar_given_tumbling_smooth)], H_sugar_given_tumbling_smooth, label='H(Sugar | Tumbling)')
axs[1, 0].set_xlabel('Time (s)')
axs[1, 0].set_ylabel('Entropy (nats)')
axs[1, 0].set_title('Sugar Entropies')
axs[1, 0].legend()

axs[1, 1].plot(time_smooth, H_sugar_derivative_smooth, label='H(Sugar Gradient)')
axs[1, 1].plot(time_smooth, H_sugar_derivative_given_methylation_smooth, label='H(Sugar Gradient | Methylation)')
axs[1, 1].plot(time_smooth, H_sugar_derivative_given_CheY_smooth, label='H(Sugar Gradient | CheY-P)')
axs[1, 1].plot(time_smooth, H_sugar_derivative_given_tumbling_smooth, label='H(Sugar Gradient | Tumbling)')
axs[1, 1].set_xlabel('Time (s)')
axs[1, 1].set_ylabel('Entropy (nats)')
axs[1, 1].set_title('Sugar Gradient Entropies')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```

```python

nbins = 30

def compute_histogram(time_series_1, time_series_2, bins):
    hist_2d, x_edges, y_edges = np.histogram2d(time_series_1, time_series_2, bins=bins)
    pxy = hist_2d / float(np.sum(hist_2d))
    px = np.sum(pxy, axis=1)
    py = np.sum(pxy, axis=0)
    return pxy, px, py

def compute_mutual_information(time_series_1, time_series_2, bins=nbins):
    pxy, px, py = compute_histogram(time_series_1, time_series_2, bins)
    nzs = pxy > 0  # Non-zero probabilities
    mi = np.sum(pxy[nzs] * np.log(pxy[nzs] / (px[:, None] * py[None, :])[nzs]))
    return mi

def concatenate_data(all_data, num_runs):
    concatenated_data = []
    for run in range(num_runs):
        concatenated_data.extend(all_data[run])
    return np.array(concatenated_data)

def compute_time_derivative(data, dt):
    derivatives = []
    for series in data:
        derivative = np.diff(series) / dt
        derivatives.append(derivative)
    return derivatives

# Assuming dt is given
dt = 0.1  # or any other value

concatenated_sugar = concatenate_data(all_sugar_concentrations, num_runs)
concatenated_methylation = concatenate_data(all_methylation_levels, num_runs)
concatenated_CheY = concatenate_data(all_CheY_levels, num_runs)
concatenated_tumbling = concatenate_data(all_tumbling_probabilities, num_runs)

time_derivative_sugar = compute_time_derivative(all_sugar_concentrations, dt)
concatenated_time_derivative_sugar = concatenate_data(time_derivative_sugar, num_runs)

length = len(concatenated_time_derivative_sugar)
concatenated_sugar = concatenated_sugar[:length]
concatenated_methylation = concatenated_methylation[:length]
concatenated_CheY = concatenated_CheY[:length]
concatenated_tumbling = concatenated_tumbling[:length]

mi_sugar_methylation = compute_mutual_information(concatenated_sugar, concatenated_methylation, bins=nbins)
mi_sugar_CheY = compute_mutual_information(concatenated_sugar, concatenated_CheY, bins=nbins)
mi_sugar_tumbling = compute_mutual_information(concatenated_sugar, concatenated_tumbling, bins=nbins)
mi_methylation_CheY = compute_mutual_information(concatenated_methylation, concatenated_CheY, bins=nbins)
mi_methylation_tumbling = compute_mutual_information(concatenated_methylation, concatenated_tumbling, bins=nbins)
mi_CheY_tumbling = compute_mutual_information(concatenated_CheY, concatenated_tumbling, bins=nbins)

mi_derivative_sugar_methylation = compute_mutual_information(concatenated_time_derivative_sugar, concatenated_methylation, bins=nbins)
mi_derivative_sugar_CheY = compute_mutual_information(concatenated_time_derivative_sugar, concatenated_CheY, bins=nbins)
mi_derivative_sugar_tumbling = compute_mutual_information(concatenated_time_derivative_sugar, concatenated_tumbling, bins=nbins)

print(f"MI(sugar, methylation): {mi_sugar_methylation}")
print(f"MI(sugar, CheY): {mi_sugar_CheY}")
print(f"MI(sugar, tumbling): {mi_sugar_tumbling}")
print(f"MI(methylation, CheY): {mi_methylation_CheY}")
print(f"MI(methylation, tumbling): {mi_methylation_tumbling}")
print(f"MI(CheY, tumbling): {mi_CheY_tumbling}")
print(f"MI(derivative_sugar, methylation): {mi_derivative_sugar_methylation}")
print(f"MI(derivative_sugar, CheY): {mi_derivative_sugar_CheY}")
print(f"MI(derivative_sugar, tumbling): {mi_derivative_sugar_tumbling}")
```
